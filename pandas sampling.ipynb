```python
import pandas as pd

# Define a function to parse the file
def parse_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = {'Title': [], 'URL': [], 'Content': []}
        current_section = None
        content = []

        for line in file:
            line = line.strip()
            if line.startswith('Title:'):
                if content:
                    data['Content'].append(' '.join(content))
                    content = []
                current_section = 'Title'
                title = line.replace('Title:', '').strip()
                data['Title'].append(title)
            elif line.startswith('URL:'):
                current_section = 'URL'
                url = line.replace('URL:', '').strip()
                data['URL'].append(url)
            elif line.startswith('Content:'):
                current_section = 'Content'
            elif current_section == 'Content' and line:
                content.append(line)

        # Add the last content block
        if content:
            data['Content'].append(' '.join(content))

    return pd.DataFrame(data)

# Path to your file
file_path = 'news_articles.txt'
df = parse_text_file(file_path)

# Save the DataFrame to a CSV file
csv_file_path = 'parsed_news_articles.csv'
df.to_csv(csv_file_path, index=False)

print("Data saved to CSV file successfully!")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Load your data
data = pd.read_csv('parsed_news_articles.csv')

# Extract features
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(data['Content'])

# Perform clustering
kmeans = KMeans(n_clusters=5, random_state=0)  # Adjust the number of clusters
data['Cluster'] = kmeans.fit_predict(X)

# Save the clustered data
data.to_csv('clustered_news_articles.csv', index=False)
```
